```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
appts_20_df = pd.read_csv("2020_morethan1.csv")
appts_20_df.head()
```

```python
appts_20_df.columns
```

```python
ind_counts_20 = appts_20_df["Individual"].value_counts().sort_index()

ind_counts_20_data = {'idx': list(ind_counts_20.index),
                      'val': list(ind_counts_20)}
```

```python
plt.bar(ind_counts_20_data['idx'], ind_counts_20_data['val'])
plt.ylabel('Frequency')
plt.xlabel('Number of Sessions')
plt.title('2020 - Number of Sessions for Individual Counseling Appts')
```

```python
group_counts_20 = appts_20_df["Group or Couple"].value_counts().sort_index()

group_counts_20_data = {'idx': list(group_counts_20.index),
                        'val': list(group_counts_20)}
```

```python
plt.bar(group_counts_20_data['idx'], group_counts_20_data['val'])
plt.ylabel('Frequency')
plt.xlabel('Number of Sessions')
plt.title('2020 - Number of Sessions for Group Counseling Appts')
```

```python
tot_counts_20 = appts_20_df["Total"].value_counts().sort_index()

tot_counts_20_data = {'idx': list(tot_counts_20.index),
                      'val': list(tot_counts_20)}
```

```python
plt.bar(tot_counts_20_data['idx'], tot_counts_20_data['val'])
plt.ylabel('Frequency')
plt.xlabel('Number of Sessions')
plt.title('2020 - Number of Sessions for Counseling Appts')
```
